#use wine data for clustering 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns ; sns.set()
```

```python
# 1. read the wine data
df_wine= pd.read_csv("../wine-labeled.csv")
```

```python
df_wine.head(5)
```

```python
# b. drop the red/white and quality columns to produce an unlabelled data set,
    # so that you don't know whether any given wine's quality or cover

df_wine= df_wine.drop(columns=["quality", "type"])
```

```python
df_wine.head(5)
```

```python
# c. using sklearn minMaxScaler scale all the data

from sklearn.preprocessing import MinMaxScaler
```

```python
# to standardize the data
# minmaxScaler will standardize the data with range from [0,1]
scaler= MinMaxScaler()
scale_data=scaler.fit_transform(df_wine)
```

```python
# d. convert the data into its first three PRINCIPAL Components
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components= 3)
```

```python
p=pca.fit_transform(scale_data)
```

```python
df_pca= pd.DataFrame(data= p, columns=["PCA_1","PCA_2", "PCA_3"])
df_pca.head(5)
```

```python
df_pca.info()
```

```python
x=np.array(df_pca["PCA_1"])
y=np.array(df_pca["PCA_2"])
y.min(), y.max()
```

```python
x.min(), x.max()
```

```python
plt.scatter(x,y)
plt.gcf().set_size_inches(10,8)
plt.grid()
plt.xlim(-0.63, 0.78)
plt.ylim(-0.60, 0.62)
```

```python
# f. Use KMeans and the silhouette Score, what are the oprimal numbers of Clusters

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X= np.array(df_pca[["PCA_1","PCA_2"]])
```

```python
model= KMeans(n_clusters= 11)
model.fit(X)
labels=model.labels_
df_pca["cluster"]= labels.reshape(-1,1)
```

```python
df_pca.head(5)
```

```python
# scatter plot to find the optimal cluster size
sc=[]
for i in range(2,10):
    model= KMeans(n_clusters= i)
    model.fit(df_pca)
    score= silhouette_score(df_pca, model.labels_)
    sc.append(score)
sc
```

```python
# use line plot and scatter plot
max_cluster=10
n_cluster= np.arange(2,max_cluster)

sns.lineplot(x= n_cluster, y=sc)
plt.scatter(x=n_cluster, y=sc, color='b')
plt.show()
```

```python
for i in range(11):
    p,q= X[ model.labels_ == i].T
    plt.scatter(p,q)
plt.gcf().set_size_inches(8,6)
```

```python
# get the pair plots


sns.pairplot(df_pca, hue='cluster')

plt.show()
```
