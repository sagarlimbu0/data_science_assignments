```python
# function to perform polynomial fits of multiple degrees from 1 to 100 errors for each degree

# this function will take, do the poly fit using the range of degrees
degree= list(range(1,100))

def poly_function(xval, yval, degree):
    xtrain, xtest, ytrain, ytest= train_test_split(xval, yval) #splits the data sets into train and Test
    
    #finding the polyfit with the xtrain
    fit= np.polyfit(xtrain, ytrain, degree)
    
    #get the y-predicted from the fit and xTest data set
    ypredict= np.polyval(fit, xtest)
    
    # get MEAN SQUARED ERROR value 
    MSE= mean_squared_error(ytest, ypredict)
    
    return (MSE)

#test with polynomial of degree 1

poly_function(xval, yval, 3), poly_function(xval, yval, 2)
```

```python
# NOW perform 100 train/test splits and collect the ERRORS as a function of degree
# you will have 100 ERRORS of each degree
degree=list(range(1,100))
# to store 100 errors of each degree, we will have mean and sigma of empty lists
means=[]
sigma=[]

for d in degree:
    errors=[]
    # iterate 
    for i in range(100):
        #using the above built-in function to store on error list
        errors.append(poly_function(xval, yval, d))
        # store the 100 errors in the mean and sigma list
    means.append(np.mean(errors))
    sigma.append(np.std(errors))
```

```python
means
```

```python
plt.plot(degree, means, marker='x')
#USING SIGMA
# first converting to numpy array

plt.plot(degree, np.array(means)+2*np.array(sigma)) # 95%
plt.plot(degree, np.array(means)-2*np.array(sigma))
#ADDING plus minus sigma
plt.fill_between(degree, np.array(means)+2* np.array(sigma), np.array(means)-2*np.array(sigma), color='pink')

# using logarithmic scale on y-axis and x-axis,
plt.xlim(0,20)
plt.ylim(0, 0.08)
#plt.yscale('linear')

plt.grid()
plt.xlabel('Degree of Fit', fontsize=16)
plt.ylabel('MSE', fontsize=16)

plt.gcf().set_size_inches(12,6)
```
