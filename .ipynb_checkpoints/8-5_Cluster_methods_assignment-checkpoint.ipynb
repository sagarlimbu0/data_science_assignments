```python
from sklearn.cluster import DBSCAN
```

```python
# create a model using DBSCAN, Density Based Clustering and Application with NOISE
```

```python
# to determine the outliers and NOISE
```

```python
model= DBSCAN( eps=.5, min_samples= 5)

model.fit(df_sm)
label= model.labels_
np.unique(label, return_counts= True)
```

```python
np.unique(model.labels_, return_counts=True)
```

```python

for i in np.array(np.unique(label)):
    if i <0:
        continue
        p,q = df_sm[ model.labels_ == i].T
        plt.scatter(p,q)
    
plt.gcf().set_size_inches(6,6)
```
